```python
import numpy as np

# Define the data
data = [
    [17, 15, 95, 43, 48],
    [44, 81, 69, 86, 95],
    [11, 18, 56, 13, 66],
    [26, 31, 22, 50, 81],
    [93, 18, 50, 55, 100],
    [18, 13, 40, 77, 84],
    [11, 52, 49, 17, 79],
    [93, 91, 71, 33, 29],
    [80, 54, 49, 90, 12],
    [77, 38, 67, 86, 100]
]

# Convert to NumPy array
array = np.array(data)

# Display the array
print(array)
```

```python

distance=[]
for i in array:
  for  j in array:
    distance.append(int(np.sqrt(int(((i-j)**2).sum()))))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Sample data: 10 rows x 5 features
data = np.array([
    [17, 15, 95, 43, 48],   # r1
    [44, 81, 69, 86, 95],   # r2
    [11, 18, 56, 13, 66],   # r3
    [26, 31, 22, 50, 81],   # r4
    [93, 18, 50, 55,100],   # r5
    [18, 13, 40, 77, 84],   # r6
    [11, 52, 49, 17, 79],   # r7
    [93, 91, 71, 33, 29],   # r8
    [80, 54, 49, 90, 12],   # r9
    [77, 38, 67, 86,100],   # r10
])

labels = ['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10']
```

```python
# Perform hierarchical/agglomerative clustering with linkage
linked = linkage(data, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, labels=labels, orientation='top', distance_sort='ascending')
plt.title("Dendrogram - Agglomerative Clustering (Ward)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.grid(True)
plt.show()
```

```python
# Agglomerative clustering (with 3 clusters)
cluster_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = cluster_model.fit_predict(data)

# Show cluster labels
df = pd.DataFrame(data, index=labels, columns=["x1", "x2", "x3", "x4", "x5"])
df["Cluster"] = clusters
print(df)
```
