```python
from gensim.models import Word2Vec
```

```python
text='''OpenAI offers two powerful third-generation embedding model \n (denoted by -3 in the model ID). Read the embedding v3 announcement blog post for more details.

Usage is priced per input token. Below is an example of pricing pages of text per US dollar \n (assuming ~800 tokens per page):'''
```

```python
text=text.splitlines()
```

```python
from gensim.utils import simple_preprocess
```

```python
tokenized_sentences=[simple_preprocess(i) for  i in text]
```

```python
model=Word2Vec(sentences=tokenized_sentences,vector_size=100,min_count=1,workers=4)
```

```python
model.wv['token']
```

```python
model.wv.most_similar("token")
```

```python
text
```

```python
model.save("word2vec_model.model")
```

```python
from sentence_transformers import SentenceTransformer

# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

# The sentences to encode
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

# 2. Calculate embeddings by calling model.encode()
embeddings = model.encode(sentences)
print(embeddings.shape)
# [3, 384]

# 3. Calculate the embedding similarities
similarities = model.similarity(embeddings, embeddings)
print(similarities)
# tensor([[1.0000, 0.6660, 0.1046],
#         [0.6660, 1.0000, 0.1411],
#         [0.1046, 0.1411, 1.0000]])
```
